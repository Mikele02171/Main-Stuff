### Task 1 - _(15 marks)_
For this task, you are to write code that computes how much you would weigh on a planet (or moon) with a different gravitational pull to Earth.

#### Instructions
Your program should ask the user for their weight on Earth and the other planet's gravitational acceleration, then print out how much they would weigh under the given circumstances.

The equation to compute the weight of an object on a planet other than Earth $w_o$ is given by:
$$
w_o = \frac{w_e}{g_e} * g_o
$$
where $w_e$ is the object's weight on Earth in kilograms; $g_e$ is Earth's gravitational acceleration; and $g_o$ is the other planet's gravitational acceleration.

Earth's gravitational acceleration has been provided as a constant. Additionally, here are a few more planets to try out.

| Celestial Body | Gravitational Acceleration (m/s^2) |
|----------------|------------------------------------|
| Sun            | 274.13                             |
| Mercury        | 3.59                               |
| Earth          | 9.81                               |
| Earth's Moon   | 1.62                               |
| Mars           | 3.77                               |
| Pluto          | 0.42                               |

_Hint: It might help to do a "sanity test" - a smaller gravitational value should result in a smaller weight._

#### Your Solution

````python
EARTH_GRAVITY = 9.81

# Write your solution here
#There are two inputs weight on earth and gravity on the other planet
w_e = float(input('Your weight on Earth (kg): '))
g_0 = float(input('Other planets gravity (m/s^2): '))

w_0 = (w_e/EARTH_GRAVITY)*g_0

# Print the weight 
print('On the other planet, you would weigh: ' + str(w_0) +' kg')
````

````python
# Write your solution here
Desposit = float(input('Deposit amount ($): '))
Interest_rate = float(input('Interest rate (%): '))
Years_invested = int(input('Years invested: '))

for b in range(Years_invested+1):
    Balance = Desposit * (((1+(Interest_rate/100)))**b)

print('Balance after 10 years: $' + str(Balance))
````
