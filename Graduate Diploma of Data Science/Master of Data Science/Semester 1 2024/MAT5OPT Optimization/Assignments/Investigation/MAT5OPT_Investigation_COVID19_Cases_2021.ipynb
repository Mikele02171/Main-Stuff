# COVID-19 US Cases 2021 

### How to convert time into time, Total Number of Cases and dates into one data-frame Excel csv file

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import glob 

#Ensure you have into the correct directory
covid19_data_2021 = glob.glob('C:/Users/Michael Le/Desktop/2021/*2021.csv')
[filename.replace('\\', '/') for filename in covid19_data_2021]
```

```python
all_dfs = []
for one_filename in [filename.replace('\\', '/') for filename in covid19_data_2021]:
    print(f'Loading {one_filename}')
    new_df = pd.read_csv(one_filename, usecols = ['Confirmed','Date'])
    all_dfs.append(new_df)
    
```

### Count the total number of days

```python
len(all_dfs)
```

```python
df = pd.concat(all_dfs)
```

```python
df
```

```python
df_no_of_cases = df.groupby('Date').sum()
df_no_of_cases 
```
