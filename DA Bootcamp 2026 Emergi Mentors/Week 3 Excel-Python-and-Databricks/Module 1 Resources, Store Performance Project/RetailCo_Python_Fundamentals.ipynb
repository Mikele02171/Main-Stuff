# 📊 PART 1: Variables - The Foundation of Data Storage

## Business Context
RetailCo's CFO sets daily revenue targets for each store. We need to store these metrics for analysis.

## What Are Variables?
Variables are named containers that store data for reuse throughout your code.

## Why This Matters
- **Reusability:** Use the same value multiple times
- **Maintainability:** Change once, updates everywhere
- **Clarity:** Name values meaningfully

## Business Impact
If the CFO changes the daily target from $50K to $55K, you update ONE variable instead of 100+ instances.

```python
# Store RetailCo's key business metrics
daily_revenue_target = 55000
company_name = "RetailCo Australia"
is_target_met = False
total_stores = 150

# Display the stored values
print("=" * 60)
print("RETAILCO AUSTRALIA - KEY BUSINESS METRICS")
print("=" * 60)
print(f"\n📊 Company: {company_name}")
print(f"🎯 Daily Revenue Target: ${daily_revenue_target:,}")
print(f"🏪 Total Stores: {total_stores}")
print(f"✅ Target Met Today: {is_target_met}")

print("\n💡 BUSINESS INSIGHT:")
print("Variables let us store and reuse values throughout our analysis.")
print("Instead of typing 50000 repeatedly, we use 'daily_revenue_target'.")
print("If the target changes, we update it ONCE, not 100 times!")
```

# 🔢 PART 2: Data Types - The Right Tool for the Right Job

## Business Context
Different business data requires different handling:
- Store names = Text
- Transaction counts = Whole numbers
- Revenue amounts = Decimal numbers
- Target achieved = Yes/No

## Python Data Types
1. **String (str):** Text data → `"Sydney CBD"`
2. **Integer (int):** Whole numbers → `1250`
3. **Float (float):** Decimal numbers → `52847.50`
4. **Boolean (bool):** True/False → `True`

## Why This Matters
Using wrong data types causes calculation errors!

```python
# RetailCo Sydney CBD Store - Different data types
store_name = "Sydney CBD"
sales_count = 1250
revenue_amount = 52847.50
target_achieved = True

print("=" * 60)
print("DATA TYPES IN ACTION - Sydney CBD Store")
print("=" * 60)

print(f"\n🏷️  Store Name: {store_name} → Type: {type(store_name).__name__}")
print(f"🔢 Sales Count: {sales_count} → Type: {type(sales_count).__name__}")
print(f"💰 Revenue: ${revenue_amount:,.2f} → Type: {type(revenue_amount).__name__}")
print(f"✅ Target Hit: {target_achieved} → Type: {type(target_achieved).__name__}")

print("\n⚠️  COMMON MISTAKE DEMONSTRATION:")
wrong_calc = '100' + '200'
right_calc = 100 + 200
print(f"Text addition: '100' + '200' = {wrong_calc}")
print(f"Number addition: 100 + 200 = {right_calc}")

print("\n💡 BUSINESS INSIGHT:")
print("Correct data types = Accurate calculations = Correct business decisions!")
```

# 📋 PART 3: Lists - Analyzing Multiple Items at Once

## Business Context
The CFO wants to see the top 10 performing stores. We need to track multiple store names and revenues together.

## What Are Lists?
Lists store multiple values in a single variable, maintaining their order.

## Why This Matters
- Analyze groups of data instead of individual items
- Perform calculations on ALL items at once
- Scale from 10 stores to 10,000 stores with same code

```python
# Top 10 RetailCo stores by daily revenue
top_10_store_names = [
    "Sydney CBD", "Melbourne Central", "Brisbane Queen St",
    "Perth CBD", "Adelaide Rundle", "Gold Coast",
    "Canberra Centre", "Newcastle", "Wollongong", "Hobart"
]

top_10_revenues = [
    52847.50, 51200.00, 49850.00, 48900.00, 47650.00,
    46800.00, 45900.00, 44750.00, 43500.00, 42300.00
]

print("=" * 60)
print("TOP 10 RETAILCO STORES - Q4 2025 PERFORMANCE")
print("=" * 60)

print(f"\n🥇 Best Store: {top_10_store_names[0]} - ${top_10_revenues[0]:,.2f}")
print(f"🥈 Second Best: {top_10_store_names[1]} - ${top_10_revenues[1]:,.2f}")
print(f"🥉 Third Best: {top_10_store_names[2]} - ${top_10_revenues[2]:,.2f}")
print(f"📊 10th Place: {top_10_store_names[9]} - ${top_10_revenues[9]:,.2f}")

print("\n📊 INSTANT ANALYSIS:")
print(f"Total Stores Analyzed: {len(top_10_store_names)}")
print(f"💰 Highest Revenue: ${max(top_10_revenues):,.2f}")
print(f"📉 Lowest in Top 10: ${min(top_10_revenues):,.2f}")
print(f"📊 Average Revenue: ${sum(top_10_revenues) / len(top_10_revenues):,.2f}")
print(f"💵 Total Revenue: ${sum(top_10_revenues):,.2f}")

print("\n💡 BUSINESS INSIGHT:")
print("Manual analysis: 10 minutes with calculator for 10 stores")
print("Python analysis: 2 seconds for ANY number of stores!")
```

# 🗂️ PART 4: Dictionaries - Real Business Data Structures

## Business Context
Real stores have multiple attributes: name, location, revenue, transactions, employees, target, manager.

## What Are Dictionaries?
Dictionaries store **key-value pairs**, connecting related information.

## Why This Matters
- This is how real databases work!
- Connect multiple pieces of information about one entity
- Access data by meaningful names

```python
# Complete profile of Sydney CBD store
sydney_cbd_store = {
    "store_name": "Sydney CBD",
    "state": "NSW",
    "daily_revenue": 52847.50,
    "transactions": 1250,
    "employees": 25,
    "target": 50000,
    "manager": "Sarah Chen",
    "opened_year": 2018
}

print("=" * 60)
print("STORE PROFILE - SYDNEY CBD")
print("=" * 60)

print(f"\n🏪 Store: {sydney_cbd_store['store_name']}")
print(f"📍 Location: {sydney_cbd_store['state']}")
print(f"👔 Manager: {sydney_cbd_store['manager']}")
print(f"👥 Staff Count: {sydney_cbd_store['employees']}")
print(f"📅 Opened: {sydney_cbd_store['opened_year']}")

print(f"\n💰 FINANCIAL PERFORMANCE:")
print(f"🎯 Daily Target: ${sydney_cbd_store['target']:,}")
print(f"💵 Actual Revenue: ${sydney_cbd_store['daily_revenue']:,.2f}")
print(f"🛒 Transactions: {sydney_cbd_store['transactions']:,}")

variance = sydney_cbd_store['daily_revenue'] - sydney_cbd_store['target']
variance_pct = (variance / sydney_cbd_store['target']) * 100
avg_transaction = sydney_cbd_store['daily_revenue'] / sydney_cbd_store['transactions']

print(f"\n📊 CALCULATED METRICS:")
print(f"{'📈' if variance > 0 else '📉'} Variance: ${variance:,.2f} ({variance_pct:+.1f}%)")
print(f"💳 Avg Transaction: ${avg_transaction:.2f}")
print(f"⚡ Revenue per Employee: ${sydney_cbd_store['daily_revenue'] / sydney_cbd_store['employees']:,.2f}")

print("\n💡 BUSINESS INSIGHT:")
print("Dictionaries = How SQL databases store records")
print("This is THE foundation for enterprise data work!")
```

# 🧮 PART 5: Calculations - Automating Business Math

## Business Context
The CFO asks these questions DAILY:
- What's our total revenue?
- What's the average per store?
- Which store performed best/worst?
- Are we meeting our targets?

## Why Python Calculations Matter
- Speed: Instant vs manual calculator work
- Accuracy: No human math errors
- Scalability: Works on 5 or 5,000 stores

```python
# Sample: 5 RetailCo stores
sample_stores = {
    "Sydney CBD": 52847.50,
    "Melbourne Central": 48200.00,
    "Brisbane Queen St": 51100.00,
    "Perth CBD": 43500.00,
    "Adelaide Rundle": 38900.00
}

print("=" * 60)
print("AUTOMATED BUSINESS CALCULATIONS - 5 STORE SAMPLE")
print("=" * 60)

total_revenue = sum(sample_stores.values())
average_revenue = total_revenue / len(sample_stores)
highest_revenue = max(sample_stores.values())
lowest_revenue = min(sample_stores.values())

print(f"\n💰 REVENUE ANALYSIS:")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Average per Store: ${average_revenue:,.2f}")
print(f"🏆 Best Performer: ${highest_revenue:,.2f}")
print(f"⚠️  Needs Attention: ${lowest_revenue:,.2f}")

revenue_range = highest_revenue - lowest_revenue
print(f"\n📊 PERFORMANCE SPREAD:")
print(f"Range: ${revenue_range:,.2f}")
print(f"Best vs Worst: {(highest_revenue/lowest_revenue - 1) * 100:.1f}% difference")

target_total = daily_revenue_target * len(sample_stores)
revenue_gap = target_total - total_revenue
achievement_rate = (total_revenue / target_total) * 100

print(f"\n🎯 TARGET PERFORMANCE:")
print(f"Combined Target: ${target_total:,}")
print(f"Combined Actual: ${total_revenue:,.2f}")
print(f"Achievement Rate: {achievement_rate:.1f}%")
print(f"{'✅ EXCEEDING' if revenue_gap <= 0 else '❌ MISSING'} by: ${abs(revenue_gap):,.2f}")

print("\n💡 BUSINESS INSIGHT:")
print("These calculations run instantly on 5 stores or 5,000 stores")
print("Manual: ~5 minutes | Python: 0.001 seconds")
```

# 🔀 PART 6: Conditional Logic - Automated Decision Making

## Business Context
RetailCo needs to categorize 150 stores automatically:
- **Excellent:** ≥110% of target → Bonus eligible
- **Good:** 100-109% of target → Maintain performance
- **Attention:** 90-99% of target → Manager coaching
- **Critical:** <90% of target → Immediate intervention

## Why This Matters
- Consistency: Same criteria applied to ALL stores
- Speed: 150 stores evaluated in seconds
- Accuracy: No human judgment errors

```python
def evaluate_store_performance(store_name, revenue, target):
    variance_pct = ((revenue - target) / target) * 100

    if revenue >= target * 1.10:
        status = "⭐ EXCELLENT"
        action = "Bonus eligible, study best practices to replicate"
    elif revenue >= target:
        status = "✅ GOOD"
        action = "On track, maintain current performance"
    elif revenue >= target * 0.90:
        status = "⚠️  ATTENTION NEEDED"
        action = "Manager coaching session required this week"
    else:
        status = "🚨 CRITICAL"
        action = "Immediate intervention - Regional Director call"

    return status, action, variance_pct

print("=" * 60)
print("AUTOMATED PERFORMANCE EVALUATION - 5 STORES")
print("=" * 60)
print(f"\nDaily Target: ${daily_revenue_target:,}\n")

for store, revenue in sample_stores.items():
    status, action, var_pct = evaluate_store_performance(store, revenue, daily_revenue_target)

    print(f"📍 {store:20} ${revenue:>10,.2f} {status}")
    print(f"   Variance: {var_pct:+6.1f}%")
    print(f"   → Action: {action}")
    print()

print("=" * 60)
print("💡 BUSINESS INSIGHT:")
print("Manual: 150 stores × 2 min = 5 hours")
print("Python: 150 stores = 2 seconds")
print("Result: Faster problem identification = Higher revenue retention")
```

# 🔁 PART 7: Loops - The Automation Superpower

## Business Context
RetailCo has 150 stores. Manually evaluating each one is time-consuming, error-prone, and expensive.

## What Are Loops?
Loops repeat actions automatically.

## Why This Matters
- Scale: Works on 1, 100, or 10,000 items
- Consistency: Exact same process for every item
- Speed: Process thousands of items per second

## Career Impact
**Loops separate data entry clerks from data analysts.**

```python
# Simulate 15 stores (imagine this is 150!)
all_stores_data = {
    "Sydney CBD": 52847, "Melbourne Central": 51200, "Brisbane QS": 49850,
    "Perth CBD": 48900, "Adelaide Rundle": 47650, "Gold Coast": 46800,
    "Canberra Centre": 45900, "Newcastle": 44750, "Wollongong": 43500,
    "Hobart": 42300, "Darwin": 38500, "Cairns": 37200,
    "Townsville": 35800, "Geelong": 34500, "Ballarat": 32100
}

print("=" * 60)
print(f"AUTOMATED ANALYSIS - {len(all_stores_data)} STORES")
print("=" * 60)
print(f"Target: ${daily_revenue_target:,}/day\n")

excellent_count = 0
good_count = 0
attention_count = 0
critical_count = 0
flagged_stores = []
total_revenue = 0

for store_name, revenue in all_stores_data.items():
    status, action, var_pct = evaluate_store_performance(store_name, revenue, daily_revenue_target)

    if "EXCELLENT" in status:
        excellent_count += 1
    elif "GOOD" in status:
        good_count += 1
    elif "ATTENTION" in status:
        attention_count += 1
        flagged_stores.append((store_name, revenue, var_pct))
    else:
        critical_count += 1
        flagged_stores.append((store_name, revenue, var_pct))

    total_revenue += revenue

print("📊 EXECUTIVE SUMMARY")
print("=" * 60)
print(f"\n⭐ Excellent Performers: {excellent_count} stores ({excellent_count/len(all_stores_data)*100:.1f}%)")
print(f"✅ Meeting Targets: {good_count} stores ({good_count/len(all_stores_data)*100:.1f}%)")
print(f"⚠️  Need Attention: {attention_count} stores ({attention_count/len(all_stores_data)*100:.1f}%)")
print(f"🚨 Critical Issues: {critical_count} stores ({critical_count/len(all_stores_data)*100:.1f}%)")

print(f"\n💰 FINANCIAL SUMMARY:")
print(f"Total Revenue: ${total_revenue:,}")
print(f"Average per Store: ${total_revenue/len(all_stores_data):,.2f}")
print(f"Combined Target: ${daily_revenue_target * len(all_stores_data):,}")

if flagged_stores:
    print(f"\n🚨 PRIORITY ACTION REQUIRED - {len(flagged_stores)} STORES:")
    print("=" * 60)
    flagged_stores.sort(key=lambda x: x[2])
    for store, revenue, var_pct in flagged_stores:
        print(f"   📍 {store:20} ${revenue:>9,} ({var_pct:+6.1f}%)")

print("\n💡 CRITICAL INSIGHT:")
print("This EXACT code works for 15, 150, or 1,500 stores")
print("You write the loop ONCE. It scales infinitely.")
print("THIS is why Python analysts earn 35% more!")
```
